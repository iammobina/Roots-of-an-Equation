```

import math
import random
import matplotlib.pyplot as plt
import numpy as np
import struct
from decimal import Decimal


pop_count = 40
mutation_rate = 0.25
crossover_rate = 0.25

# initial population
def initializePopulation(size):
    population = []
    for i in range(size):
        chromosome = []
        for i in range(10):
            gene = random.randint(0, 1)
            chromosome.append(gene)
        population.append(chromosome)
    return population


# Convertor
def convertor(binary):
    y = 0
    for j in binary:
        y = str(y) + str(j)
    # return  decimal.Decimal(str(y))
    # print(y)
    # float_value=float(y)
    # print(float_value)
    # return Decimal(float_value)
    return float(int(y,2))


def expression(x):
    y =9*(x**5) - 194.7*(x**4.0) + 1680.1*(x**3.0) - 7227.94*(x**2.0) + 15501.2*x - 13257.2
    return y


def fitness_func(chromosomes):
    fitness_list = []
    for chromosome in chromosomes:
        chrom_fitness = expression(convertor(chromosome))
        fitness_list.append(chrom_fitness)


    fitness_list=np.array(fitness_list)

    fitness_list=(min(fitness_list)-np.absolute(fitness_list)) -fitness_list
    return [x for (y, x) in sorted(zip(fitness_list, chromosomes), reverse=1)], sorted(fitness_list, reverse=1)


# crossover
def crossover(p1, p2):
    child = p1[0:3] + p2[3:10]
    return child


# bit switch
def switch(gene):
    if gene == 1:
        gene = 0
    elif gene == 0:
        gene = 1
    return gene


def mutate(chromosome):
    for index in range(len(chromosome)):
        if (random.random() < mutation_rate):
            chromosome[index] = switch(chromosome[index])
    return chromosome


X = range(300)
Y = []
for i in range(300):
    Y.append(expression(i))
population = initializePopulation(pop_count)
chromosomes, fitness_list = fitness_func(population)

for i in range(2000):
    print("generation: " + str(i), "chromosome: " + str(chromosomes[0]), "X: " + str(convertor(chromosomes[0])))
    chromosomes = chromosomes[0:int(pop_count/2)]

    while(len(chromosomes)<pop_count):
        if random.random() < crossover_rate:
            parents = random.sample(range(0, len(chromosomes)-1), 2)
            parent1 = parents[0]
            parent2 = parents[1]
            child = crossover(chromosomes[parent1], chromosomes[parent2])

            if random.random() < mutation_rate:
                child = mutate(child)
            chromosomes.append(child)


    chromosomes, fitness_list = fitness_func(chromosomes)

print("Root is :" ,convertor(chromosomes[0]))




```
